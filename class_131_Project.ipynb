```python
import pandas as pd
import csv

df = pd.read_csv("https://raw.githubusercontent.com/Akriti96/class130_Project/main/finalData.csv")
df.head()
```

```python
df.drop(["Unnamed: 0"], axis = 1, inplace = True)
df
```

```python
#df.dtypes

#df['Radius'] = df['Radius'].astype(float, errors = raise)
df['Radius'] = pd.to_numeric(df['Radius'], errors = "coerce")
df['Distance'] = pd.to_numeric(df['Distance'], errors = "coerce")
df['Mass'] = pd.to_numeric(df['Mass'], errors = "coerce")
df.dtypes
```

```python
df1 = df.dropna(how='any',axis=0)
df1
```

```python
df1.reset_index(inplace=True)
df1
```

```python
radius = df1['Radius'].tolist()
mass = df1['Mass'].tolist()
def converttoSIunit(mass, radius):
  for i in range(0,len(radius)):
    radius[i] = radius[i]*6.957e+8
    mass[i] = mass[i]*1.989e+30

converttoSIunit(mass, radius)
print(mass)
print(radius)
```

```python
Gravity = []
def GravityCal(mass, radius):
  G = 6.674e-11
  for i in range(0, len(radius)):
    g = (mass[i]*G)/(radius[i]**2)
    Gravity.append(g)

GravityCal(mass, radius)
print(Gravity)
```

```python
df1['Gravity'] = Gravity
df1
```
